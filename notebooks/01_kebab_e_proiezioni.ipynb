# Popolazione, Kebab e invecchiamento

Questo foglio di lavoro analizza la struttura della popolazione italiana: livello totale, distribuzione per età e sesso, quote per grandi classi, età media, mediana, quantili e dipendenza degli anziani.

Il **Kebab demografico** usa una visualizzazione formalmente nota come piramide demografica. Qui è usato per leggere il peso delle coorti senza dare per scontata una base larga.

Fonti principali: [`demo_pjan`](https://ec.europa.eu/eurostat/databrowser/view/demo_pjan/default/table?lang=en), [`proj_23np`](https://ec.europa.eu/eurostat/databrowser/view/proj_23np/default/table?lang=en), [`migr_pop3ctb`](https://ec.europa.eu/eurostat/databrowser/view/migr_pop3ctb/default/table?lang=en). Ogni grafico riporta fonte ed elaborazione in basso a sinistra.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
while ROOT != ROOT.parent and not (ROOT / "pyproject.toml").exists():
    ROOT = ROOT.parent

if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from demografia.notebook_charts import (
    fig_age_distribution,
    fig_age_shares,
    fig_births_deaths,
    fig_dependency,
    fig_education_distribution,
    fig_education_trend,
    fig_europe_rank,
    fig_europe_series,
    fig_fertility,
    fig_kebab,
    fig_kebab_animation,
    fig_migration,
    fig_population_series,
    fig_regional_rank,
    fig_regional_series,
    load_notebook_tables,
    notebook_paths,
)

paths = notebook_paths(ROOT)
tables = load_notebook_tables(paths["final"])
```

## Parametri

I valori sotto controllano l'intero foglio. Per cambiare paese di confronto usare codici ISO3 con prefisso `country:`; per togliere il confronto usare `none`.

```python
PAESE = "country:ITA"
PAESE_CONFRONTO = "country:ESP"

ANNO_OSSERVATO = 2024
ANNO_PROIEZIONE = 2041
ANNO_NATI_ESTERO = 2025

ANNO_INIZIO_EVOLUZIONE = 1960
ANNO_FINE_EVOLUZIONE = 2050
PASSO_EVOLUZIONE = 1
```

## Serie storica della popolazione

La prima lettura controlla la dimensione complessiva della popolazione osservata e proiettata. Il confronto predefinito è la Spagna, utile perché evita la discontinuità tedesca precedente al 1991.

```python
fig_population_series(tables, territory=PAESE, compare=PAESE_CONFRONTO, metric="population_total").show()
```

## Kebab osservato

La forma per età e sesso mostra quali coorti pesano di più nell'anno osservato. Uomini a sinistra, donne a destra.

```python
fig_kebab(tables, territory=PAESE, year=ANNO_OSSERVATO).show()
```

## Kebab proiettato

La stessa lettura può essere spostata sugli anni di proiezione. Quando dati osservati e proiettati si sovrappongono, vengono preferiti gli anni osservati.

```python
fig_kebab(tables, territory=PAESE, year=ANNO_PROIEZIONE).show()
```

## Evoluzione anno per anno

Il grafico animato parte dal primo anno disponibile e permette di seguire lo spostamento delle coorti nel tempo. Il passo è impostato a un anno, ma può essere aumentato per alleggerire l'animazione.

```python
fig_kebab_animation(
    tables,
    territory=PAESE,
    start_year=ANNO_INIZIO_EVOLUZIONE,
    end_year=ANNO_FINE_EVOLUZIONE,
    step=PASSO_EVOLUZIONE,
).show()
```

## Popolazione nata all'estero

Questa vista isola lo stock di popolazione residente nata all'estero, distinto dalla cittadinanza. La copertura parte dal 2002 nella fonte usata.

```python
fig_kebab(tables, territory=PAESE, year=ANNO_NATI_ESTERO, population_kind="foreign_born").show()
```

## Grandi classi di età

Le quote 0-14, 15-64, 65+ e 80+ sintetizzano la redistribuzione tra popolazione giovane, adulta e anziana.

```python
fig_age_shares(tables, territory=PAESE, compare=PAESE_CONFRONTO).show()
```

## Età media, mediana e quantili

Media e mediana non raccontano la stessa cosa: la media sintetizza tutta la distribuzione, mentre mediana e quantili mostrano dove si collocano centro e code.

```python
fig_age_distribution(tables, territory=PAESE, compare=PAESE_CONFRONTO).show()
```

## Dipendenza demografica

Il tasso di dipendenza degli anziani mette in rapporto la popolazione 65+ con la popolazione 15-64. La dipendenza totale include anche la componente 0-14.

```python
fig_dependency(tables, territory=PAESE, compare=PAESE_CONFRONTO).show()
```
