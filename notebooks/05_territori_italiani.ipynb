Fonti principali: [`demo_r_pjangroup`](https://ec.europa.eu/eurostat/product?code=demo_r_pjangroup&language=EN&mode=view), [`demo_r_gind3`](https://ec.europa.eu/eurostat/databrowser/view/demo_r_gind3/default/table?lang=en), [`demo_r_find3`](https://ec.europa.eu/eurostat/databrowser/view/demo_r_find3/default/table?lang=en). Ogni grafico riporta fonte ed elaborazione in basso a sinistra.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
while ROOT != ROOT.parent and not (ROOT / "pyproject.toml").exists():
    ROOT = ROOT.parent

if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from demografia.notebook_charts import (
    fig_age_distribution,
    fig_age_shares,
    fig_births_deaths,
    fig_dependency,
    fig_education_distribution,
    fig_education_trend,
    fig_europe_rank,
    fig_europe_series,
    fig_fertility,
    fig_kebab,
    fig_kebab_animation,
    fig_migration,
    fig_population_series,
    fig_regional_rank,
    fig_regional_series,
    load_notebook_tables,
    notebook_paths,
)

paths = notebook_paths(ROOT)
tables = load_notebook_tables(paths["final"])
```
