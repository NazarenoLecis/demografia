## Parametri

Le fasce disponibili includono `15-64`, `25-64`, `25-34`, `35-44`, `45-54` e `55-64`. Il sesso `T` indica totale, `M` uomini, `F` donne.

```python
PAESE = "ITA"
PAESE_CONFRONTO = "ESP"

FASCIA_PRINCIPALE = "25-64"
FASCIA_GIOVANI = "25-34"
SESSO = "T"
ANNO = 2024
```

## Distribuzione per livello di istruzione

Il grafico mostra come si distribuisce la popolazione nella fascia selezionata tra i principali livelli ISCED. Il confronto è modificabile dai parametri.

Ordine di lettura: dal livello pi? basso al terziario. La voce *secondaria generale* corrisponde a ISCED 34 e 44; la voce *secondaria professionale* corrisponde a ISCED 35 e 45. Entrambe sono sottocategorie della secondaria superiore e post-secondaria non terziaria. La voce *fino alla licenza media* ? l'aggregato ISCED 0-2: in questa fonte Eurostat non ? possibile separare primaria/elementare e secondaria inferiore.

```python
fig_education_distribution(tables, country=PAESE, compare=PAESE_CONFRONTO, age_label_value=FASCIA_PRINCIPALE, sex=SESSO, year=ANNO).show()
```

```python
fig_education_distribution(tables, country=PAESE, compare=PAESE_CONFRONTO, age_label_value=FASCIA_GIOVANI, sex=SESSO, year=ANNO).show()
```
