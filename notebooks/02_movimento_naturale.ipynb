Ogni grafico riporta fonte ed elaborazione in basso a sinistra.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
while ROOT != ROOT.parent and not (ROOT / "pyproject.toml").exists():
    ROOT = ROOT.parent

if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from demografia.notebook_charts import (
    fig_age_distribution,
    fig_age_shares,
    fig_births_deaths,
    fig_dependency,
    fig_education_distribution,
    fig_education_trend,
    fig_europe_rank,
    fig_europe_series,
    fig_fertility,
    fig_kebab,
    fig_kebab_animation,
    fig_migration,
    fig_population_series,
    fig_regional_rank,
    fig_regional_series,
    load_notebook_tables,
    notebook_paths,
)

paths = notebook_paths(ROOT)
tables = load_notebook_tables(paths["final"])
```

## Parametri

I territori nazionali usano `country:ISO3`; regioni e province italiane usano codici NUTS con prefisso `region:` o `province:`.

```python
PAESE = "country:ITA"
PAESE_CONFRONTO = "country:ESP"

REGIONE = "region:ITC4"      # Lombardia
REGIONE_CONFRONTO = "region:ITI4"  # Lazio
PROVINCIA = "province:ITC4C" # Milano
PROVINCIA_CONFRONTO = "province:ITC11" # Torino

ANNO_TERRITORI = 2024
```

## Fecondità e natalità

La fecondità totale misura i figli per donna; il tasso di natalità esprime i nati per 1.000 abitanti. Letti insieme separano intensità riproduttiva e dimensione della popolazione.

```python
fig_fertility(tables, territory=PAESE, compare=PAESE_CONFRONTO).show()
```

## Nati, decessi e saldo naturale

Il saldo naturale è la differenza tra nati vivi e decessi. La serie storica aiuta a distinguere cali strutturali da shock di breve periodo.

```python
fig_births_deaths(tables, territory=PAESE, compare=PAESE_CONFRONTO).show()
```

## Serie dedicate ai singoli indicatori

## Lettura regionale e provinciale

Le classifiche ordinano i territori nell'ultimo anno disponibile o nell'anno indicato nei parametri. Le serie permettono di verificare un territorio specifico senza perdere la dinamica temporale.

```python
fig_regional_rank(tables, level="region", metric="live_births", year=ANNO_TERRITORI, limit=25).show()
```

```python
fig_regional_rank(tables, level="province", metric="live_births", year=ANNO_TERRITORI, limit=40).show()
```

```python
fig_regional_series(tables, focus=REGIONE, compare=REGIONE_CONFRONTO, metric="live_births").show()
```

```python
fig_regional_series(tables, focus=PROVINCIA, compare=PROVINCIA_CONFRONTO, metric="deaths").show()
```
