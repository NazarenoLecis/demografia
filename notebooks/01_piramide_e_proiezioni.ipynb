# Piramide demografica e proiezioni


```python
from pathlib import Path
import pandas as pd
from demografia.charts import plot_population_pyramid

ROOT = Path.cwd().resolve()
while ROOT != ROOT.parent and not (ROOT / "demografia").exists():
    ROOT = ROOT.parent
population = pd.read_parquet(ROOT / "output/data/final/population_age_sex_observed_projected.parquet")
```

```python
years = [year for year in (1990, 2024, 2050, 2080) if year in set(population.loc[population.iso3.eq("ITA"), "year"])]
for year in years:
    display(plot_population_pyramid(population, "ITA", year, ROOT / f"output/charts/piramide_italia_{year}.png"))
```
